```python
# 1.query() - query function which only requires to write the condition for filtering.

df.query('x < y')
```

```python
# 2.insert() - if we want to add a new column to a dataframe 

new_colm=np.zeros(53940,dtype=int)

df.insert(2,'new_colm',new_colm)  #(pos,colname,object)
df
```

```python
# 3.sample() - select values randomly from DataFrame.

df.sample() # it gives only one random value...if we want 2 obs. put n=2 inside bracket
```

```python
# 4.loc and iloc - Loc and iloc are used to select rows and columns.

# 4.1 loc - select by labels

# Selecting first 3 rows and first 2 columns with loc:

df.loc[:2,['carat','cut']]
```

```python
# 4.2 iloc - select by position

# Selecting first 3 rows and first 2 columns with iloc:

df.iloc[:3,:2]
```

```python
# 5.where() - it works is that values that fit the condition are selected 

df.where(df['x']<4)
```

```python
# 6.rank() - Rank function assigns rank to the values.

df['rank_1']=df['carat'].rank()
df
```

```python
# 7.nunique() - nunique counts the number of unique entries over columns or rows

df.nunique()  # if we want only one col then df.carat.nunique()
```

```python
# 7.1 unique() - it will return unique elements of particular column

df.cut.unique()
```

```python
# 8.memory usage() - it will returns how much memory used by columns in bytes

df.memory_usage()
```

```python
# 9.rename() -  to rename columns name permanently...pass a dictionary

df.rename(columns={'price':'prices'},inplace=True)
df
```

```python
# 10.drop() - # To drop any column temporaray...for permanent put (inplace=True)

df.drop(['color'],axis=1)
```

```python
# 11.replace() - it replaces the values of a column...for permanent put (inplace=True)

df.replace(3.95,4)
```

```python
# 11.1
# If we need to change more than one value, 
# we can pass a dictionary to the replace function 
# where the key should be the original value and the value should be the replacement.
df.replace({61.5:61,59.8:60,56.9:60})
```

```python
# 12.value_counts() - We can get the value counts of each category

df['color'].value_counts()
```

```python
# 13.crosstab() - It gives a frequency table

pd.crosstab(df['cut'],df['color']) # we got the number count of all the combinations of cut and color
```

```python
# 14.transpose() - in this rows become columns and the columns become rows.

df.transpose()
```

```python
# 15.duplicated() - it gives no of duplicate rows

df.duplicated().sum()
```

```python
# 16.corr() - This function is used to find the correlation of all columns in the dataframe df.corr()

df.corr()
```

```python
# 17.sort_values() - it sort all values of given column in ascending order

df.sort_values('depth',ascending=False) # it gives values in descending order
```

```python
# 18.max() - it gives maximum value of each column...if you wants maximum values of rows just put axis=1

df.max()
```

```python
# 18.1 min() - it gives minimum value of each column...if you wants minimum values of rows just put axis=1


df.min()
```

```python
# 19.clip() - it will clips or replace out-of-range values in given range

df['prices'].clip(300,400)
```

```python
# 20.at() - it tells us at given index and column which values is there

df.at[3,'cut']  # (index,colname)
```

```python
# 20.1 iat() 

df.iat[3,1]  # (index,index)
```

```python
# 3.distribution plot

x=df['carat']
plt.title("Distribution chart")
sns.distplot(x,hist=True,kde=True,color='orange')
sns.set_style("darkgrid")
plt.show()
```

```python
# 4.bar chart
plt.figure(figsize=(10,5))
df['clarity'].value_counts().plot(kind='bar',color='maroon')

plt.show()
```
